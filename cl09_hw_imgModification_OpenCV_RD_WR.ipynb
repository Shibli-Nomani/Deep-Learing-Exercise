```python

def ImageDataLoader(Number_of_images):
    imagePath = glob.glob( "../content/drive/MyDrive/DeepLearning-AIQUEST/openCV_image/sample/dogs/*.jpg")
    cv_img = []

    for img in imagePath:
    
        img = cv2.imread(img)
        img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    
        cv_img.append(img)
        #print(cv_img)
        
    #to plot 5X5 images
    fig = plt.figure(figsize=(15,15))
    for i in range(0,25):
        fig.add_subplot(5,5, i+1)
        plt.imshow(cv_img[i])
```

**Define Funcation to show Image**

```python
def ImageShow(imageTitle, image):
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    plt.imshow(image)
    plt.title(imageTitle)
    plt.grid(False)
    plt.tight_layout()
    plt.show()
```

**Movement in image**

```python
image_path = {"Image": "../content/drive/MyDrive/DeepLearning-AIQUEST/openCV_image/sample/dogs/dog.31.jpg"}
imageData = cv2.imread(image_path['Image'])
imageData
```

**To check output by calling the custom function called ImageShow**

```python
def ImageShow(imageTitle, image):
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    plt.imshow(image)
    plt.title(imageTitle)
    plt.grid(False)
    plt.tight_layout()
    plt.show()
```

```python
movementInfo= np.float32([[1, 0, 50], [0, 1, 100]]) #width =50, height = 70
updatedImageAfterMovement = cv2.warpAffine(imageData, movementInfo, (imageData.shape[1], imageData.shape[0]))
ImageShow("updated Image After Movement", updatedImageAfterMovement)
```

```python
movementInfo = np.float32([[1, 0, -50], [0, 1, -70]]) #width =-50, height = -70
updatedImageAfterMovement = cv2.warpAffine(imageData, movementInfo, (imageData.shape[1], imageData.shape[0]))
ImageShow("updated Image After Movement", updatedImageAfterMovement)
```
